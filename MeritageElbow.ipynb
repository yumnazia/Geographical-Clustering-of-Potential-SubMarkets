```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline


df = pd.read_excel('C:/Meritage Project/Merged File with VDL information.xlsx') #Reading the Data file
df.head()
```

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
```

```python
model = KMeans()
visualizer = KElbowVisualizer(model, k=(3,12))

visualizer.fit(df)    # Running the Elbow method to find the optimal number of clusters
visualizer.poof() 
```

```python
from sklearn.cluster import KMeans

kmeans=KMeans(n_clusters=5)    #Doing the Kmeans for 5 Clusters
kmeans.fit(df)

kmeans.labels_
```

```python
x=kmeans.predict(df)
x
```

```python
df['new_val']=x
```
